```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
resource_path = os.getcwd() + "/resource/"
```

```python
img = cv2.imread(resource_path + 'dsu1.jpg',1)
rows,cols,ch = img.shape
pts1 = np.float32([[319,276],[761,313],[764,508],[331,536]])
pts2 = np.float32([[1,1],[443,2],[443,193],[2,193]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(443,193))
plt.figure(figsize=(25,20))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img2 = cv2.imread(resource_path + 'dsu1.jpg',1)
```

```python
gray = cv2.imread(resource_path + 'dsu1.jpg',0)
edges = cv2.Canny(gray,50,200,apertureSize = 3)
# minLineLength = 300
# maxLineGap = 10
lines = cv2.HoughLinesP(edges,1,np.pi/180,1,minLineLength = 150,maxLineGap = 10)
# lines = cv2.HoughLines(edges, 1, np.pi / 180, 1)
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(img2,(x1,y1),(x2,y2),(0,255,0),2)
```

```python
plt.figure(figsize=(25, 20))
plt.subplot(121),plt.imshow(edges),plt.title('Input')
plt.subplot(122),plt.imshow(img2),plt.title('Output')
plt.show()
```
